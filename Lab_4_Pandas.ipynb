```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
plt.hist(data["Age"], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#The histogram shows the distribution of ages in the dataset. It helps identify the most frequent age group and understand how the data is spread.
```

```python
plt.figure(figsize=(8,5))
plt.hist(data["Likes"], bins=10)
plt.title("Likes Distribution")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Frequency plots help visualize how many times each category appears. They make it easy to compare categorical data.
```

```python
#Why Visualization is Important?
#Data visualization helps understand patterns, trends, and relationships in data. It makes complex data easy to interpret and helps identify distributions, outliers, and frequency of values.
```
